## Measurement Pads

```python
JJn       = 14 # number of test junction spaces
JJspacing = 25 # spacing from junction to junction >> width of the inner divs
Gspacing  = 5 # spacing to ground   >> width of the outer divs
C1width   = 5

Slength   = 100 # length of the measurement lines starter

r         = 570 # distance between starter and measurement pad

Mlength   = 140 # length of the measurement pad
Mwidth    = 75 # width of the measurement pad
Mspace    = 55 # spacing between the measurement pads

off = 5
alpha     = (JJn+off)*(2*Mwidth+Mspace+JJspacing)/(2.*pi*r) # degree of the triangle; 

Rsmall = JJn * JJspacing + 2*Gspacing # small rectangle size
```

```python
JJn       = 2 # number of test junction spaces
JJspacing = 25 # spacing from junction to junction >> width of the inner divs
Gspacing  = 5 # spacing to ground   >> width of the outer divs
C1width   = 5

Slength   = 50 # length of the measurement lines starter

r         = 75 # distance between starter and measurement pad

Mlength   = 90 # length of the measurement pad
Mwidth    = 75 # width of the measurement pad
Mspace    = 55 # spacing between the measurement pads

off = 5
alpha     = (JJn+off)*(2*Mwidth+Mspace+JJspacing)/(2.*pi*r) # degree of the triangle; 

Rsmall = JJn * JJspacing + 2*Gspacing # small rectangle size
```

```python
JJboxlength = 50
JJboxwidth = 20

sum_length = Slength + r + Mlength + JJboxlength # overall length 

P0 = LineString([(0,0),(1,0)])

print P0.xy

Pboundary = SD.empty()
Pstructure = []

############ for RS estimation ###############
xRS= {i:[] for i in arange(JJn)}
yRS = {i:[] for i in arange(JJn)}
PRS = {i:[] for i in arange(JJn)}

WHOLE_SHELL = []
ppoints = []

for i in arange(JJn):
    
    # starters on the left
    P1 = scale(P0,
               xfact=Slength,
               origin=(0,0))
    
    P1 = translate(P1,
                   yoff=i*Rsmall/JJn)
    
    
    # space for the junctions  
    
    PJ = scale(P0, 
                xfact=-JJboxlength,
                origin=(0,0))
    
    PJ = translate(PJ, 
                    xoff=P1.xy[0][0],
                    yoff=P1.xy[1][0])
    
    if not i%2:
        Ppatch = scale(P0,
               xfact=-JJboxlength,
               origin=(0,0))
        
        Ppatch = translate(Ppatch,
                           xoff=P1.xy[0][0],
                   yoff=P1.xy[1][0]+JJboxwidth/2.)
        
        ppoints += [[P1.xy[0][0], P1.xy[1][0]]]
        
        Ppatch2 = scale(P0,
               xfact=JJspacing,
               origin=(0,0))
        
        Ppatch2 = rotate(Ppatch2,
               angle = 90,
               origin=(0,0))
        
        Ppatch2 = translate(Ppatch2,
                           xoff=P1.xy[0][0],
                           yoff=P1.xy[1][0])
    
    # ... and the last line on the right (the measurement pad)
    
    P3 = scale(P0, 
               xfact=Mlength, #+ abs(sin(i*pi/JJn)*Mlength), #(JJn/2-i)
               origin=(0,0))
    
    P3 = translate(P3,
                   xoff=Slength+r, 
                   yoff=i*(Mwidth+Mspace/2)-Mwidth/2.) #Rsmall/2.+JJspacing+(i-JJn/2.)*(Mwidth+Mspace/2)) #+P1.xy[1][-1] +
    
    # for the second line: connect the starter to the last line
    
    P2 = Polygon([(P1.xy[0][-1],P1.xy[1][-1]+C1width/2),
                  (P1.xy[0][-1],P1.xy[1][-1]-C1width/2),
                  (P3.xy[0][0],P3.xy[1][0]-Mwidth/2),
                  (P3.xy[0][0],P3.xy[1][0]+Mwidth/2)])
    
    # give the starter and beginning line the right width
    P11 = P1.buffer(C1width/2,cap_style=3)
    P33 = P3.buffer(Mwidth/2,cap_style=2)
    PJJ = PJ.buffer(JJboxwidth/2, cap_style=3)
    
    # union of the shapes:
    Pfinal = P33.union(P2)
    Pfinal = Pfinal.union(P11)
    
    
    # buffer to get the groundplane spacing
    Pfunny = Pfinal.buffer(Gspacing)
    
    Pboundary = Pboundary.union(Pfunny)
    Pboundary = Pboundary.union(PJJ)
    
    if i%2:
        #Ppatch2 = Ppatch2.buffer(C1width/2., cap_style=3)
        #Pfinal = Pfinal.union(Ppatch2)
        
        Ppatch = Ppatch.buffer(JJboxwidth/2., cap_style=3)
        Pboundary = Pboundary.union(Ppatch)
    
    ########### RS estimation 
    xRS[i] += list(P1.xy[0])
    yRS[i] += list(P1.xy[1])
    xRS[i] += list(P3.xy[0])
    yRS[i] += list(P3.xy[1])
    PRS[i] = Pfinal
    
    Pstructure.append(Pfinal)
    
Pstructure = SD.unary_union(Pstructure)
Pstructure2 = scale(Pstructure, xfact = -1)
Pstructure2 = translate(Pstructure2, xoff = -sum_length)
Pstructure2 = list(SD.flattenMultipoly(Pstructure2))
Pstructure = list(SD.flattenMultipoly(Pstructure))
Pstructure = Pstructure + Pstructure2

Pboundary2 = scale(Pboundary, xfact = -1)
Pboundary2 = translate(Pboundary2, xoff = -sum_length)
Pboundary = SD.flattenMultipoly(Pboundary)
Pboundary2 = SD.flattenMultipoly(Pboundary2)

Pboundary = list(Pboundary) + list(Pboundary2)

Pstructure
```

### Define function:

```python
#### loop ####

w_lo = 10 
h_lo = 6.5 # outer loop height
ih_lo = 3.5   # inner loop height  

#### launcher1 ####

w_l1 = w_lo
h_l1 = 20

#### junctions ####

#w_jl = 0.714
#w_jr = 1.071
w_b  = 0.4

#### undercut box ####
w_u = 0.5

#### horizontal connection bar ###
wbar = 45
hbar = 8

### writefield ###
writefield = 500

def make_testsquidarray(Nsquids, w_jl, w_jr):
                   
    l1 = SD.mybox((0,0), w_l1, h_l1)
    
    l1 = translate(l1, yoff=-h_l1)
    # anker points to attach the SQUID loop
    xdummy, y_l1_anker = SD.anker(l1,
                                  'lower_left')
    x_l1_anker, ydummy = SD.anker(l1,
                                  'center')
    
    l2 = SD.mybox((0,0), w_l1, h_l1)
    
    l2 = translate(l2, yoff=-h_l1)
    # anker points to attach the SQUID loop
    xdummy, y_l2_anker = SD.anker(l2,
                                  'upper_left')
    x_l2_anker, ydummy = SD.anker(l2,
                                  'center')
    
    ############ SQUID LOOP ############
    ####################################
    
    ######### OUTER LOOP ############
    
    lo = SD.mybox((0,0), w_lo, h_lo)
    xdummy, y_lo_anker = SD.anker(lo,
                                  'upper_left')
    x_lo_anker, ydummy = SD.anker(lo,
                                  'center')
    
    #plot(x_lo_anker,y_lo_anker,'o')
    #SD.showPolygons([lo])

    ######### INNER LOOP ############
    
    loi   = SD.mybox((0,0), w_lo-w_jl-w_jr, ih_lo)
    y_loi = (h_lo-ih_lo)/2.
    x_loi = w_jl
    loi   = translate(loi, xoff=x_loi,
                            yoff = y_loi)
    
    #SD.showPolygons([loi])
    
    ######### BRIDGE ############
    
    lob   = SD.mybox((0,0), w_lo, w_b)
    y_lob = h_lo/2. - w_b/2.
    lob   = translate(lob, yoff=y_lob)
    #SD.showPolygons([lob])

    ######## UNDERCUT BOXES ###########
    ub    = SD.mybox((0,0), w_lo-w_jl-w_jr, w_u)
    x_ub  = w_jl
    y_ub1 = (h_lo - ih_lo)/2.

    y_ub2 = y_ub1 + ih_lo - w_u
    
    ub1 = translate(ub, xoff = x_ub,
                        yoff = y_ub1)
    
    ub2 = translate(ub, xoff = x_ub,
                        yoff = y_ub2)
    
    
    ######################################
    ######### FINALIZE & PLOT ############
    ######################################
    
    squid_loop = lo.difference(loi)
    squid_loop = squid_loop.difference(lob)
    squid_loop = SD.flattenMultipoly(squid_loop)
    #SD.showPolygons(squid_loop)
    squid_loop = list(squid_loop) #+ list([ub1,ub2])
    
    SQUID_loop = SD.BRAND_NEW_STRUCTURE(SD.empty(),[])
    SQUID_loop.EBL.add_polis(squid_loop)
    #SQUID_loop.add_EBL(squid_loop)
    SQUID_loop.add_anker([x_lo_anker, y_lo_anker],'upper center')
    SQUID_loop.add_anker([x_lo_anker, 0],'lower center')
    
    SQUID_loop.add_layer([ub1, ub2], 'UNDERCUT')
    
    xl, yl = SQUID_loop.ANKERS['upper center']
    x1, y1 = x_l1_anker, y_l1_anker
    
    SQUID_loop.translate([x1,y1],
                         [xl,yl])

    dummy, yl1u = SD.anker(l1, 'upper_left')
    xl1u, dummy = SD.anker(l1, 'center')
    
    lbar = SD.mybox((0,0),wbar,hbar)
    lbarx, dummy = SD.anker(lbar,'center')
    dummy, lbary = SD.anker(lbar,'lower_left')

    TESTSQUIDS = SD.BRAND_NEW_STRUCTURE([],[])
    TESTSQUIDS.add_layer([],'UNDERCUT')
    
    figure(figsize=(10,10))

    lbar1 = translate(lbar, xoff=xl1u-lbarx, yoff=yl1u-lbary)
    
    TESTSQUIDS.add_anker([xl1u,yl1u],'mid')

    TESTSQUIDS.EBL.add_polis([lbar1,l1])
    
    TESTSQUIDS.EBL.add_polis(SQUID_loop.get_polygons('EBL'))
    TESTSQUIDS.UNDERCUT.add_polis(SQUID_loop.get_polygons('UNDERCUT'))
    
    nsquids = Nsquids-1
    
    ox, oy = SQUID_loop.ANKERS['lower center']

    
    for i in arange(nsquids):
        SL = SQUID_loop.make_copy()
        nx, ny = SL.ANKERS['upper center']
        SL.translate([ox,oy], [nx,ny])
        ox, oy = SL.ANKERS['lower center']
        
        TESTSQUIDS.EBL.add_polis(SL.get_polygons('EBL'))
        TESTSQUIDS.UNDERCUT.add_polis(SL.get_polygons('UNDERCUT'))
           
    lend = translate(l2, xoff=ox-x_l2_anker, yoff=oy-y_l2_anker)
    lbar2 = translate(lbar, xoff=xl1u-lbarx, yoff=oy-y_l2_anker-h_l1-hbar)
    TESTSQUIDS.EBL.add_polis([lend, lbar2])
    TESTSQUIDS.add_anker([xl1u, hbar/2.], 'upper center')
    
    P = TESTSQUIDS.get_polygons('EBL')
    xy = SD.unary_union(P).centroid.xy
    x0, y0 = xy[0][0], xy[1][0]
    TESTSQUIDS.add_anker([x0, y0], 'the center')
    
    WF = SD.mybox((-writefield/2., -writefield/2.), writefield, writefield)
    WF = SD.translate(WF, xoff = x0, yoff = y0) # translate the WF to the center of the SQUID
    TESTSQUIDS.add_layer([WF], 'WF')
    
    return TESTSQUIDS
```

```python
ts = [0.8, 0.9, 1.0, 1.1, 1.2, 2.25]
Rexp = 325
w1opt = 0.714
w2opt = 1.071
offset = array([51.5,0])

rese = []
resu = []

SQUIDS_asymm = []

for i, t in enumerate(ts):
    TS = make_testsquidarray(5, t*w1opt, t*w2opt)
    print t*w1opt, t*w2opt, t*w1opt+t*w2opt
    SQUIDS_asymm += [TS]
    TS.translate(i*offset, TS.ANKERS['mid'])
    e = TS.get_polygons('EBL')
    u = TS.get_polygons('UNDERCUT')
    rese += e
    resu += u
    
    
SD.showPolygons(rese+resu)
grid()
SD.mouseshow()
```
